```python
# CARD
# SUIT,RANK,VALUE

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}

# DECK

import random

# PLAYER
```

```python
class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
```

```python
class Deck:
    
    def __init__(self):
        
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                # create the card object
                created_card = Card(suit,rank)
                
                self.all_cards.append(created_card)
                
    def shuffle(self):
        
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        
        return self.all_cards.pop(0)
```

```python
class Player:
    
    def __init__(self,name):
        
        self.name = name
        self.all_cards = []
        
    def remove_one(self):
        return self.all_cards.pop(0)
    
    def add_cards(self,new_cards):
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)
        else:
            self.all_cards.append(new_cards)
    
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards.'
```

```python
player1 = Player('Taz')
player2 = Player('Sam')

game_deck = Deck()
game_deck.shuffle()

for i in range(26):
    player1.add_cards(game_deck.deal_one())
    player2.add_cards(game_deck.deal_one())
```

```python
game_on = True
round_num = 0

while game_on:
    
    round_num += 1
    print(f'Round {round_num}')
    
    if len(player1.all_cards) <= 6:
        print(f'{player1.name} out of cards')
        game_on = False
        break
        
    if len(player2.all_cards) <= 6:
        print(f'{player2.name} out of cards')
        game_on = False
        break
        
    # New Round
    tablecards = []
    tablecards.append(player1.remove_one())
    tablecards.append(player2.remove_one())
    
    war = True
    while war:
        if tablecards[0].value > tablecards[1].value:
            random.shuffle(tablecards)
            player1.add_cards(tablecards)
            print(f'{player1.name} wins round {round_num} and has {len(player1.all_cards)} cards!')
            war = False
            break
        elif tablecards[0].value < tablecards[1].value:
            random.shuffle(tablecards)
            player2.add_cards(tablecards)
            print(f'{player2.name} wins round {round_num} and has {len(player2.all_cards)} cards!')
            war = False
            break
        else:
            for i in range(5):
                tablecards.append(player1.remove_one())
                tablecards.append(player2.remove_one())
            tablecards.insert(0,player1.remove_one())
            tablecards.insert(1,player2.remove_one())
            print(f'WAR! There are {len(tablecards)} cards on the table.')

    # while war
```
